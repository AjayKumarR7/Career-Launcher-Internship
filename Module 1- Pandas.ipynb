### Query 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def prices_func(N):
    day = max(dataset['Date']) - pd._libs.tslibs.timedeltas.Timedelta(str(N-1) + 'days' + ' 00:00:00')
    index = True
    for i in range(0, len(dataset)):
        if(day == dataset['Date'][i]):
            index = i
            break
    if(index == True):
        day = max(dataset['Date']) - pd._libs.tslibs.timedeltas.Timedelta(str(N) + 'days' + ' 00:00:00')
        for i in range(0, len(dataset)):
            if(day == dataset['Date'][i]):
                index = i
                break
    if(index == True):
        day = max(dataset['Date']) - pd._libs.tslibs.timedeltas.Timedelta(str(N + 1) + 'days' + ' 00:00:00')
        for i in range(0, len(dataset)):
            if(day == dataset['Date'][i]):
                index = i
                break
    prices = []
    for i in range(index, len(dataset)):
        prices.append(dataset['Close Price'][i])
    return(prices)
```

```python
def mean_func(prices):
    mean_price_n = sum(prices)/len(prices)
    return(mean_price_n)
```

```python
def profit_or_loss_func(prices):
    profit_or_loss = ((prices[-1] - prices[0])/prices[0])*100
    return(profit_or_loss)
```

```python
to_find = ['1 week', '2 weeks', '1 month', '3 months', '6 months', '1 year']
```

```python
day = []
for i in range(0, len(to_find)):
    if('day' in to_find[i]):
        day.append(int(to_find[i][0] + to_find[i][1]))
    if('week' in to_find[i]):
        day.append(int(to_find[i][0] + to_find[i][1]) * 7)
    if('month' in to_find[i]):
        day.append(int(to_find[i][0] + to_find[i][1]) * 30)
    if('year' in to_find[i]):
        day.append(int(to_find[i][0] + to_find[i][1]) * 365)
```

```python
print(day)
```

```python
mean_prices = []
profit_or_loss_perct = []
for i in range(0, len(day)):
    N = day[i]
    prices = prices_func(N)
    mean_prices.append(to_find[i] + ': ' + str(round(mean_func(prices), 2)))
    profit_or_loss_perct.append(to_find[i] + ': ' + str(round(profit_or_loss_func(prices), 2)))
```

```python
print(mean_prices)
```

```python
print(profit_or_loss_perct)
```

### Query 1.7
Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5 <br>
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1 <br>
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1 <br>
'Positive' for 'Day_Perc_Change' in between 1 and 3 <br>
'Negative' for 'Day_Perc_Change' in between -1 and -3 <br>
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7 <br>
'Among top losers' for 'Day_Perc_Change' in between -3 and -7 <br>
'Bull run' for 'Day_Perc_Change' >7 <br>
'Bear drop' for 'Day_Perc_Change' <-7 <br>

```python
trend = []
for i in range(0, len(dataset['Day_Perc_Change'])):
    if((dataset['Day_Perc_Change'][i] < 0.5) & (dataset['Day_Perc_Change'][i] >= -0.5)):
        trend.append('Slight or No change')
    if((dataset['Day_Perc_Change'][i] < 1) & (dataset['Day_Perc_Change'][i] >= 0.5)):
        trend.append('Slight positive')
    if((dataset['Day_Perc_Change'][i] < -0.5) & (dataset['Day_Perc_Change'][i] >= -1)):
        trend.append('Slight negative')
    if((dataset['Day_Perc_Change'][i] < 3) & (dataset['Day_Perc_Change'][i] >= 1)):
        trend.append('Positive')
    if((dataset['Day_Perc_Change'][i] < -1) & (dataset['Day_Perc_Change'][i] >= -3)):
        trend.append('Negative')
    if((dataset['Day_Perc_Change'][i] < 7) & (dataset['Day_Perc_Change'][i] >= 3)):
        trend.append('Among top gainers')
    if((dataset['Day_Perc_Change'][i] < -3) & (dataset['Day_Perc_Change'][i] >= -7)):
        trend.append('Among top losers')
    if(dataset['Day_Perc_Change'][i] >= 7):
        trend.append('Bull run')
    if(dataset['Day_Perc_Change'][i] < -7):
        trend.append('Bear drop')
```

```python
dataset['Trend'] = pd.DataFrame(trend)
```
