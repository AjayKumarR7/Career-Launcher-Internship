### Query 4.1a 

Create a new column 'Call' , whose entries are - 

>'Buy' if the stock price is below the lower Bollinger band 

>'Hold Buy/ Liquidate Short' if the stock price is between the lower and middle Bollinger band 

>'Hold Short/ Liquidate Buy' if the stock price is between the middle and upper Bollinger band 

>'Short' if the stock price is above the upper Bollinger band




```python
call = []
for i in range(0, len(dataset)):
    if(dataset['Average Price'][i] <= dataset['Lower_Band'][i]):
        call.append('Buy')
    if((dataset['Average Price'][i] > dataset['Lower_Band'][i]) & (dataset['Average Price'][i] <= dataset['14_Day_Rolling_Mean'][i])):
        call.append('Hold Buy/ Liquidate Short')
    if((dataset['Average Price'][i] > dataset['14_Day_Rolling_Mean'][i]) & (dataset['Average Price'][i] <= dataset['Upper_Band'][i])):
        call.append('Hold Short/ Liquidate Buy')
    if(dataset['Average Price'][i] > dataset['Upper_Band'][i]):
        call.append('Short')
```

```python
dataset['Call'] = pd.DataFrame(call)
```

```python
dataset
```

### Query 4.1b

Now train a classification model with the 3 bollinger columns and the stock price as inputs and 'Calls' as output. Check the accuracy on a test set. (There are many classifier models to choose from, try each one out and compare the accuracy for each)

```python
X = dataset.iloc[:, [8, 18, 19, 20]].values
y = dataset.iloc[:, 21].values

from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy = ' + str(accuracy) + '%')
```

```python
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy = ' + str(accuracy) + '%')
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy = ' + str(accuracy) + '%')
```

```python
# K Nearest Neighbours

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy = ' + str(accuracy) + '%')
```

```python
# Support Vector Machine

from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy = ' + str(accuracy) + '%')
```

It can be seen that Support Vector Machine is the best model for this problem giving an accuracy of 97.58%
