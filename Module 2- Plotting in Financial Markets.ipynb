### Query 2.4 
We had created a Trend column in module 1. We want to see how often each Trend type occurs. This can be seen as a pie chart, with each sector representing the percentage of days each trend occurs. Plot a pie chart for all the 'Trend' to know about relative frequency of each trend. You can use the groupby function with the trend column to group all days with the same trend into a single group before plotting the pie chart. From the grouped data, create a BAR plot of average & median values of the 'Total Traded Quantity' by Trend type. 

```python
labels = np.unique(np.array(dataset['Trend']))
labels
```

```python
length = []
for i in range(0, len(labels)):
    length.append(len(dataset.groupby('Trend').get_group(labels[i])))
```

```python
length
```

```python
sum(length)
```

```python
patches, texts = plt.pie(length, startangle=90, radius=1.2)
labels_legend = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(labels, length)]
plt.legend(labels_legend, loc='left center', bbox_to_anchor=(-0.1, 1.),fontsize=8)
plt.show()
```

```python
mean = []
median = []
for i in range(0, len(labels)):
    mean.append(np.mean(np.array(dataset.groupby('Trend').get_group(labels[i])['Total Traded Quantity'])))
    median.append(np.median(np.array(dataset.groupby('Trend').get_group(labels[i])['Total Traded Quantity'])))
```

```python
mean
```

```python
median
```

```python
fig = plt.subplots(figsize=(20,10))
plt.bar(labels, mean)
plt.xlabel('Trend')
plt.ylabel('Mean')
plt.show()
```

```python
fig = plt.subplots(figsize=(20,10))
plt.bar(labels, median)
plt.xlabel('Trend')
plt.ylabel('Median')
plt.show()
```

### Query 2.9
Trade Calls - Using Simple Moving Averages. Study about moving averages here. 
 
Plot the 21 day and 34 day Moving average with the average price and decide a Call ! 
Call should be buy whenever the smaller moving average (21) crosses over longer moving average (34) AND the call should be sell whenever smaller moving average crosses under longer moving average. 
One of the most widely used technical indicators.

```python
# Since Saturdays and Sundays are holidays, we are going to take the rolling averages with a window of 15 and 25.
```

```python
labels = ['Average Price', '21_SMA', '34_SMA']
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.plot( dataset.index.values, dataset['Average Price'])
ax.plot( dataset.index.values, dataset['Average Price'].rolling(15).mean())
ax.plot( dataset.index.values, dataset['Average Price'].rolling(25).mean())
for i in range(1, len(dataset['Average Price'].rolling(15).mean()) - 1):
    if((dataset['Average Price'].rolling(15).mean()[i] > dataset['Average Price'].rolling(25).mean()[i]) & (dataset['Average Price'].rolling(15).mean()[i-1] < dataset['Average Price'].rolling(25).mean()[i-1])):
        ax.plot( dataset.index.values[i], dataset['Average Price'].rolling(15).mean()[i], marker = 'v', markerfacecolor = 'black', markersize = 12)
    if((dataset['Average Price'].rolling(15).mean()[i] < dataset['Average Price'].rolling(25).mean()[i]) & (dataset['Average Price'].rolling(15).mean()[i-1] > dataset['Average Price'].rolling(25).mean()[i-1])):
        ax.plot( dataset.index.values[i], dataset['Average Price'].rolling(15).mean()[i], marker = 'v', markerfacecolor = 'green', markersize = 12)
fig.autofmt_xdate()
spacing = 43
visible = ax.xaxis.get_ticklabels()[::spacing]
for label in ax.xaxis.get_ticklabels():
    if label not in visible:
        label.set_visible(False)
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend(labels, bbox_to_anchor=(-0.1, 1.),fontsize=8)
plt.show()
```

### Query 2.10 
Trade Calls - Using Bollinger Bands 
Plot the bollinger bands for this stock - the duration of 14 days and 2 standard deviations away from the average 
The bollinger bands comprise the following data points- 
>The 14 day rolling mean of the closing price (we call it the average) <br>

>Upper band which is the rolling mean + 2 standard deviations away from the average.<br> 

>Lower band which is the rolling mean - 2 standard deviations away from the average.<br> 

>Average Daily stock price.<br>

Bollinger bands are extremely reliable , with a 95% accuracy at 2 standard deviations , and especially useful in sideways moving market. 
Observe the bands yourself , and analyse the accuracy of all the trade signals provided by the bollinger bands. 
Save to a new csv file. 

```python
# Since Saturdays and Sundays are holidays, we are going to take the rolling average with a window of 10.
```
